```python
img = cv2.imread("/content/download.jpeg")
```

```python
imgBlue = img[:,:,0]
imgGreen = img[:,:,1]
imgRed = img[:,:,2]

new_img = np.hstack((imgBlue, imgGreen, imgRed))
cv2_imshow(new_img)
```

```python
# resize
img_resize = cv2.resize(img, (256, 256))
cv2_imshow(img_resize)
```

```python
# Fliping image
img_flip = cv2.flip(img, 1) # 0, 1, -1
cv2_imshow(img_flip)
```

```python
img_flip = cv2.flip(img, 0) # 0, 1, -1
cv2_imshow(img_flip)
```

```python
img_flip = cv2.flip(img, -1) # 0, 1, -1
cv2_imshow(img_flip)
```

```python
# image crop
img_crop = img[100:200, 200:300]
cv2_imshow(img_crop)
```

```python
img_crop = img[56:200, 30:45]
cv2_imshow(img_crop)
```

```python
img_crop = img[0:100, 0:100]
cv2_imshow(img_crop)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# drawing shape and twxt on image
img = np.zeros((512, 512, 3))

# rectangle
cv2.rectangle(img, pt1=(384, 0), pt2=(510, 128), color=(0, 255, 0), thickness=3)

# circle
cv2.circle(img, center=(256, 256), radius=100, color=(0, 0, 255), thickness=3)

# line
cv2.line(img, pt1=(0, 0), pt2=(511, 511), color=(255, 0, 0), thickness=3)

# text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, text="OpenCV", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=3)

print(img.shape)
cv2_imshow(img)
```

```python
# drawing shape and twxt on image
img = np.zeros((512, 512, 3))

# rectangle filled
cv2.rectangle(img, pt1=(384, 0), pt2=(510, 128), color=(0, 255, 0), thickness=-1)

print(img.shape)
cv2_imshow(img)
```
